### A code block to read and sort the electricity spot pricing from Elering between 01.04.2025 00:00:00 - 28.02.2026 23:45:00.

### The end result is a properly sorted CSV file.

```python
import pandas as pd
```

```python
#Reading the CSV file
df = pd.read_csv("spot_market_price_apr25_to_mar26_raw.csv", sep = ";", decimal = ",")
```

```python
#Renaming the columns
df = df.rename(columns={
    'Periood': 'timestamp',
    'Börsihind senti/kWh (sisaldab käibemaksu)': 'spot_price_incl_vat',
    'Börsihind senti/kWh (ei sisalda käibemaksu)': 'spot_price_excl_vat'
})
```

```python
#Changing the "timestamp" format from DD.MM.YYYY HH:MM to DD.MM.YYYY HH:MM:SS
df['timestamp'] = pd.to_datetime(df['timestamp'], format='%d.%m.%Y %H:%M').dt.strftime('%d.%m.%Y %H:%M:%S')
```

```python
#Splitting the timestamp column into date and time columns while keeping the original timestamp
df[['date','time']] = df.timestamp.str.split(" ",expand=True)
```

```python
#Converting the prices from cents/kwh to EUR/kwh
df[['spot_price_incl_vat', 'spot_price_excl_vat']] = df[['spot_price_incl_vat', 'spot_price_excl_vat']] / 100
```

```python
#Removing the FIRST duplicates from the daylight saving time date
df = df.drop_duplicates(
    subset=['timestamp'],
    keep='last'
)
```

```python
#Removing the last row as there is no consumption on that timestamp
df.drop(df.tail(1).index,
        inplace = True)
```

```python
#Rearranging the columns
df = df[['timestamp', 'date', 'time', 'spot_price_incl_vat', 'spot_price_excl_vat']]
```
